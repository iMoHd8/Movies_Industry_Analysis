# Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
matplotlib.rcParams['figure.figsize'] = (15,8)
```

# Reading The Data

```python
df = pd.read_csv('movies.csv')
```

```python
df.isna().sum()
```

```python
df.head(50)
```

## Filling the Null Values with the mean of that Column

```python
for col in df.columns:
    if col == 'budget' or col == 'gross':
        if df['budget'].isna and df['gross'].isna:
            df['budget'].fillna(0)
            df['gross'].fillna(0)
    if df[col].dtype == 'float' or df[col].dtype == 'int':
        df[col] = df[col].fillna(int(df[col].mean()))
```

```python
df.head(50)
```

## Converting budget, gross, and votes Columns to Integer Data Type

```python
df['budget'] = df['budget'].astype(int)
df['gross'] = df['gross'].astype(int)
df['votes'] = df['votes'].astype(int)
```

## Extractint the Correct Release Year From 'released' Column and Assigning it in New Column 'released_year'

```python
df['released_year'] = df['released'].str.extract(pat= "([0-9]{4})").fillna(0).astype(int)
df[['released_year']]
```

## Re-sorting the Columns and Showing the New Sorted-Columns Data

```python
df = df.reindex(sorted(df.columns), axis=1)
df.head()
```

## Sorting the Data Based on Gross ascendingly 

```python
df = df.sort_values(by=['gross'],ascending=False)
df.head()
```

## Creating The Revenue Column Based on Gross and Budget Columns

```python
df['revenue'] = df['gross'] - df['budget']
df.head()
```

# Starting the Visualizations and Extracting the Correlations

## Scatter Figure that Shows the Correlation Between Budget and Gross

```python
plt.scatter(x='budget', y='gross', data=df, color = 'darkred', alpha=0.5)
plt.title("Budget VS Gross gainess")
plt.xlabel("Budget")
plt.ylabel("Gross")
plt.ylim(0)
plt.grid(False)
```

## Regression Figure that Shows the Correlation Between Budget and Votes using seaborn

```python
sns.regplot(x='budget', y='votes', data=df, line_kws={"color":"blue"}, scatter_kws={"color":"gray"})
plt.title("Budget VS Votes")
plt.xlabel("Budget")
plt.ylabel("Votes")
plt.ylim(0)
plt.gca().set_facecolor('#F5F5F5')
plt.grid(False)
```

## The Correlation Table of Numerical Columns

```python
df.corr(method="pearson")
```

## Heat-Map of the Correlations 

```python
correlations_metrix = df.corr(method="pearson")
sns.heatmap(correlations_metrix, annot=True)
plt.title("The Correlation Metric for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
```

## Copying the Data to Another DataFrame to Convert the Object Values into Numerical Values'
### Each Object Value (String Value) Converted to Unique Numerical Value

```python
categorical_df = df.copy()
for col in categorical_df.columns:
    if categorical_df[col].dtype == 'object':
        categorical_df[col] = categorical_df[col].astype('category')
        categorical_df[col] = categorical_df[col].cat.codes
categorical_df.head(10)
```

## Creating a New Correlation Table with The New Numerical Values

```python
categorical_df.corr(method="pearson")
```

## The New Heat-Map of the Correlations

```python
all_data_correlations_metrix = categorical_df.corr(method="pearson")
sns.heatmap(all_data_correlations_metrix, annot=True, cmap='vlag')
plt.title("The Correlation Metric for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
```

# Top Countries Based on Budget

```python
top_countries = df.groupby('country')['budget'].sum().sort_values(ascending=False).head(5)
bars = plt.bar(top_countries.index, top_countries.values, color=['#9BABB8', '#EEE3CB', '#D7C0AE', '#967E76', '#A0C49D'])
plt.xlabel('Country')
plt.title('The Most 5 Countries by Budget')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, '{:,.2f}'.format(height), ha='center', va='bottom')
plt.gca().set_facecolor('white')
plt.grid(False)
```

# Displaying Top 5 Companies by Gross Income

```python
top_companies = df.groupby('company')['gross'].sum().sort_values(ascending=False).head(5)
bars = plt.bar(top_companies.index, top_companies.values, color=['#78C1F3', '#9BE8D8', '#E2F6CA', '#F8FDCF', '#FDFFAE'])
plt.xlabel('Company')
plt.title('Top 5 Companies by Gross Income')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, '{:,.2f}'.format(height), ha='center', va='bottom')
plt.gca().set_facecolor('white')
plt.grid(False)
```

# Displaying Top 5 Companies by Rating

```python
top_companies_by_score = df.groupby('company')['score'].mean().sort_values(ascending=False).head(5)
bars = plt.bar(top_companies_by_score.index, top_companies_by_score.values, color=['#071952', '#0B666A', '#35A29F', '#97FEED', '#78C1F3'])
plt.xlabel('Company')
plt.title('Top 5 Companies With The Highest Score')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, '{:,.1f} / 10'.format(height), ha='center', va='bottom')
plt.gca().set_facecolor('darkgray')
plt.grid(False)
```
